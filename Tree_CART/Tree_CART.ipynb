```python
def split_up(data, index, value):
    left_group, right_group = list(), list()
    for row in data:
        if row[index] < value:
            left_group.append(row)
        else:
            right_group.append(row)
    return left_group, right_group
```

```python
def gini_criterion(groups, classes):
    total_samples = float(sum(len(group) for group in groups))
    gini_index = 0
    for group in groups:
        group_size = float(len(group))
        if group_size == 0:
            continue
        proportion_square = .0
        for class_index in classes:
            proportion = [row[-1] for row in group].count(class_index) / len(group)
            proportion_square += pow(proportion, 2)
        gini_index += (1 - proportion_square) * (group_size / total_samples)
    return gini_index
```

```python
import math 

def information_gain(groups, classes):
    total_samples = float(sum([len(group) for group in groups]))
    source_group = []
    for group in groups:
        source_group += group
    S0 = .0
    for class_index in classes:
        S0 -= ([row[-1] for row in source_group].count(class_index) / len(source_group))*math.log2([row[-1] 
                                            for row in source_group].count(class_index) / len(source_group))
    IG = S0
    for group in groups:
        group_size = float(len(group))
        if group_size == 0:
            continue
        Si = .0
        for class_index in classes:
            if([row[-1] for row in group].count(class_index) / len(group)) == 0:
                continue
            Si -= ([row[-1] for row in group].count(class_index) / len(group))*math.log2([row[-1] 
                                                for row in group].count(class_index) / len(group))
        IG -= Si*(group_size / total_samples)
        
    return IG
```

```python
def best_split(data, flg):
    class_values = list(set(row[-1] for row in data))
    best_index, best_value, best_score_gini, best_score_ig, best_groups = 9999, 9999, 9999, -1, None
    for index in range(len(data[0]) - 1):
        for row in data:
            groups = split_up(data, index, row[index])
            
            if flg:
                gini = gini_criterion(groups, class_values)
                if gini < best_score_gini:
                    best_index, best_value, best_score_gini, best_groups = index, row[index], gini, groups
            
            else:
                inf_gain = information_gain(groups, class_values)
                if inf_gain > best_score_ig:
                    best_index, best_value, best_score_ig, best_groups = index, row[index], inf_gain, groups
    
    return {'index':best_index, 'value':best_value, 'score_gini': best_score_gini, 
            'score_ig': best_score_ig, 'groups':best_groups}
                
```

```python
def terminal(group):
    y = [row[-1] for row in group]
    return max(set(y), key=y.count)
```

```python
def rec(data, depth, criterion): 
    left, right = data['groups']
    del(data['groups'])
    
    if not left or not right:
        data['left'] = data['right'] = terminal(left + right)
        return
    if data['score_gini'] == 0:
        data['left'] = terminal(left)
        data['right'] = terminal(right)
        return
    if data['score_ig'] == 0:
        data['left'] = terminal(left)
        data['right'] = terminal(right)
        return
    if not depth:
        data['left'] = terminal(left)
        data['right'] = terminal(right)
        return
    if len(left) <= 10:
        data['left'] = terminal(left)
    else:
        data['left'] = best_split(left, criterion)
        rec(data['left'], depth - 1, criterion)
    if len(right) <= 10:
        data['right'] = terminal(right)
    else:
        data['right'] = best_split(right, criterion)
        rec(data['right'], depth - 1, criterion)
    print('Split: [X%d < %.3f] Gini = %.3f; IG = %.3f' % ((data['index'] + 1), data['value'],
                                                         data['score_gini'], data['score_ig']))
    print('Depth = ', depth)
```

```python
def tree(data, max_depth, criterion):
    root = best_split(data, criterion)
    rec(root, max_depth, criterion)
    return root
```

```python
def predict(data, row):
    if row[data['index']] < data['value']:
        if isinstance(data['left'], dict):
            return predict(data['left'], row)
        else:
            return data['left']
    else:
        if isinstance(data['right'], dict):
            return predict(data['right'], row)
        else:
            return data['right']
```

```python
def print_tree(data, depth=0):
    if isinstance(data, dict):
        print('%s[X%d < %.3f]' % ((depth*' ', (data['index']+1), data['value'])))
        print_tree(data['left'], depth+1)
        print_tree(data['right'], depth+1)
    else:
        print('%s[%s]' % ((depth*' ', data)))
```

```python
main_tree = tree(data[:46], 8, True)
```

```python
pred = []
true = []
for row in data[105:]:
    prediction = predict(main_tree, row)
    pred.append(int(prediction))
    true.append(row[-1])
# print(pred)
# print(true)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(true, pred)
```
